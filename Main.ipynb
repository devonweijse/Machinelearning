```python
import pandas as pd
import matplotlib as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.metrics import accuracy_score as accuracy
%matplotlib inline
```

```python
data = pd.read_csv("creditcard.csv") 
test = data[:100]
test = test[["Time", "Amount", "V1", "V2", "V3", "Class"]]
```

```python
print(test)
```

```python
print("Time")
print("Mean              ", test.Time.mean())
print("Standard Deviation", test.Time.std())
print("Median            ", test.Time.median())
print("Standard Error    ", test.Time.sem())
print("")

print("Amount")
print("Mean              ", test.Amount.mean())
print("Standard Deviation", test.Amount.std())
print("Median            ", test.Amount.median())
print("Standard Error    ", test.Amount.sem())
print("")

print("V1")
print("Mean              ", test.V1.mean())
print("Standard Deviation", test.V1.std())
print("Median            ", test.V1.median())
print("Standard Error    ", test.V1.sem())
print("")

print("V2")
print("Mean              ", test.V2.mean())
print("Standard Deviation", test.V2.std())
print("Median            ", test.V2.median())
print("Standard Error    ", test.V2.sem())
print("")

print("V3")
print("Mean              ", test.V3.mean())
print("Standard Deviation", test.V3.std())
print("Median            ", test.V3.median())
print("Standard Error    ", test.V3.sem())
print("")
```

```python
fig = plt.pyplot.figure()
ax = Axes3D(fig)
ax.scatter(test.Time, test.Amount, test.Class,)
```
